## Running the polarity score on the entrie dataset

```python
ar.df
```

```python
df = ar.df
```

```python
res={}
for i,row in tqdm(df.iterrows(),total=len(df)):
    text=row['Text']
    myid=row['Id']
    res[myid]=sia.polarity_scores(text)
```

```python
res
```

Storing in DataFrame as it is easy to work with

```python
pd.DataFrame(res)
```

```python
#flipping data horizontally
pd.DataFrame(res).T
```

```python
vaders= pd.DataFrame(res).T
vaders=vaders.reset_index()\
.rename(columns={'index':'Id'})
#one column name shoulb be same for merge
```

```python
vaders
```

#### Megring in original data frame

```python
vaders= vaders.merge(df,how='left')
```

```python
vaders
```

```python
#Now we have sentiment score and metadata 
vaders.head()
```

```python
ax=sns.barplot(data=vaders,x='Score',y='compound')
ax.set_title('Compound Score By Amazon Review')
plt.show()



```

```python
fig, axs=plt.subplots(1,3,figsize=(15,5))
sns.barplot(data=vaders,x='Score',y='pos',ax=axs[0])
sns.barplot(data=vaders,x='Score',y='neu',ax=axs[1])
sns.barplot(data=vaders,x='Score',y='neg',ax=axs[2])
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')
plt.show()
```
